## Preprocessing

```python
# Split del dataset para entrenamiento y test:
X = df.iloc[:, df.columns != 'price']
y = df['price']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.15, random_state=42)
```

```python
# Busqueda de hiperparametros optimos:
model = neighbors.KNeighborsRegressor()
param_grid = {'n_neighbors':range(5,35),
              'weights':['uniform','distance']}
gs = model_selection.GridSearchCV(estimator = model,
                                  param_grid = param_grid)
result = gs.fit(X_train, y_train)
```

```python
# Score de Cross-validation:
result.best_score_
```

```python
# Mejor combinación de parametros:
result.best_params_
```
