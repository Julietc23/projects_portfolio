```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
```

## CARGA DE DATOS

```python
data = load_iris()
df = pd.DataFrame(data.data, columns = data.feature_names)
df['target'] = data.target
df
```

## PREPROCESAMIENTO DE DATOS

```python
# Split de datos
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=42)
```

```python
# Elbow method para la inicialización de KMeans:

inertia = []
for n in range(1,10):
    model = KMeans(n_clusters=n)
    model.fit(X_train, y_train)
    inertia.append(model.inertia_)
```

```python
# Busqueda grafica del numero de centroides ideal:
plt.figure(figsize=(8,6))

plt.scatter(range(1,10), inertia, c = 'r')
plt.plot(range(1,10), inertia, c = 'y')
plt.xlabel('Número de centroides')
plt.ylabel('Valor de inercia')
plt.title('Inercia según número de centroides')
plt.show()
```

## MODELADO

```python
# Modelado
clf = KMeans(n_clusters=3)
clf.fit(X_train, y_train)
```

```python
df['pred'] = clf.predict(data.data)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,8))

ax1.scatter(df["sepal length (cm)"], df["sepal width (cm)"], c =  df.target)
ax1.set_title("Iris class")

ax2.scatter(df["sepal length (cm)"], df["sepal width (cm)"], c = df.pred)
ax2.scatter(clf.cluster_centers_[0][0],clf.cluster_centers_[0][1], linewidths = 3, marker='D', c = 'b', label='cent 1')
ax2.scatter(clf.cluster_centers_[1][0],clf.cluster_centers_[1][1], linewidths = 3, marker='D', c = 'r', label='cent 2')
ax2.scatter(clf.cluster_centers_[2][0],clf.cluster_centers_[2][1], linewidths = 3, marker='D', c = 'g', label='cent 3')
ax2.set_title("Cluster ID")
ax2.legend() 
```
