## Preprocessing of data

```python
# Dado que tengo muchas variables, conviene realizar una reducción de dimensiones con MCA:
X = df.iloc[:, df.columns != 'y']
mca = prince.MCA(n_components=3,
    n_iter=3,
    copy=True,
    check_input=True,
    engine='sklearn',
    random_state=42
)
mca = mca.fit(X)
df1 = mca.transform(X)
df1
```
