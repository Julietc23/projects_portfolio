```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## Creación de datos

```python
# Reproductibilidad
np.random.seed(0)

# Parametros de la función
a = np.random.uniform(0,10)
b = np.random.uniform(0,10)

# Generación de datos
x = np.linspace(1, 10, 100)

# Función:
y = b + a*np.log(x)

# Función con ruido:
intensidad = 2
ruido = np.random.normal(1, intensidad, x.shape)
y_ruidoso = y + ruido
```

## Visualización

```python
plt.scatter(x, y_ruidoso, label='Función logaritmica ruidosa', c='r')
plt.plot(x, y, label='Función logaritmica sin ruido', c = 'b')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Distribución de datos con base logaritmica')
plt.legend()
plt.show()
```

## Entrenamiento

```python
# Parametro para el entrenamiento
log_x = np.log(x.reshape(-1, 1))

# Ajuste
adj_model = linear_model.LinearRegression().fit(log_x, y)
```

```python
# Comparación de coeficientes
a_result = adj_model.coef_[0]
b_result = adj_model.intercept_

print(f"""Coef. linealizado : {a_result}
Coef. no linealizado : {a}""")
```

```python
y_pred = adj_model.predict(log_x)
```

```python
plt.scatter(x, y, label="Datos Originales", color="red", alpha=0.5)
plt.plot(x, y_pred, label=f"Ajuste Logarítmico: y = {a_result:.2f} * log(x) + {b_result:.2f}", color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Ajuste de Modelo Logarítmico")
plt.show()
```
