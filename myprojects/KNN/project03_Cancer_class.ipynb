```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import model_selection, neighbors, preprocessing
%matplotlib inline
```

## Carga de datos:

```python
df = pd.read_csv('C:/Users/agust/Desktop/project portfolio/myprojects/KNN/datasets/Cancer_data.csv')
df
```

## Exploración y limpieza de datos:

```python
# Existencia de valores faltantes:
df.isnull().sum()
```

```python
# Limpieza:
df.dropna(axis=1, inplace=True) #decido esto ya que se trata de un error en la carga, no representa información alguna.
```

```python
df.info()
```

```python
# debido a que la columna de 'id' no representa información relevante al modelo:
df.drop(labels='id', inplace=True, axis=1)
df
```

```python
df.describe()
```

## Preprocessing:

```python
# Binarización con one hot encoding de la variable target 'diagnosis':
label_enc = preprocessing.LabelEncoder()
df['diagnosis'] = label_enc.fit_transform(df['diagnosis'])
df
```

```python
# Split de dataset:
X = df.iloc[:, df.columns != 'diagnosis']
y = df['diagnosis']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
# Busqueda de hiperparametros optimos con GSCV:

clf = neighbors.KNeighborsClassifier()
param_grid = {"n_neighbors" : range(5,31),
              "weights" : ['uniform','distance']
             }

gs = model_selection.GridSearchCV(estimator = clf,
                                  param_grid = param_grid,
                                  scoring = None,
                                  cv = None)
result = gs.fit(X_train, y_train)
```

```python
# Mejor configuración:
result.best_params_
```

```python
# Score (accuracy):
result.best_score_
```

## Visualización de los datos:

```python
df.columns
```

```python
cols = ["radius",'texture','perimeter','area','smoothness','compactness','concavity','concave points','symmetry','fractal_dimension']
```

```python
def gral_heatmap(col):
    # columna de media por cada caracteristica
    df[col+"_summary"] = df[[col+'_mean', col+'_se', col+'_worst']].mean(axis=1)
    
    # Matriz de corr
    corr_matrix = df[[col+'_mean', col+'_se', col+'_worst', col+"_summary", "diagnosis"]].corr()
    
    # Ploteo del heatmap
    plt.figure(figsize=(8, 5))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f')
    plt.title(f'HeatMap para {col}')
    plt.show()
```

```python
for col in cols:
    gral_heatmap(col)
```

```python
analyzed = ["radius_worst", "texture_worst", "perimeter_worst", "area_summary", "smoothness_summary", "compactness_summary", "concavity_summary", "concave points_summary", "symmetry_summary", "fractal_dimension_worst", "diagnosis"]
df_analyzed = df[analyzed]
```

```python
for col in df_analyzed:
    sns.histplot(data = df_analyzed, x = col, hue = "diagnosis")
    plt.show()
```
